## Quick Access - Sponsor activity insights and Reports

This Notebook is to quick access and prepare Individual sponsor level insights. In the following list of Sponsors included within the time span of 2005 to 2025, pick any one to get their reports after running this Notebook. 

**Note:** The Insights and Charts will be saved in the corresponding path of this Jupyter Notebook and saved in png format.


```python
from Root import generate_sponsor_report, generate_model_driven_sponsor_dashboard
import pandas as pd
# load the raw dataset
df = pd.read_csv("Model_ready_df.csv")

SPONSOR_NAME = input(" Enter Sponsor Name > ") # Type any fro the following list of Sponsors
generate_sponsor_report(
    df,
    SPONSOR_NAME,
    save_path=f"{SPONSOR_NAME.replace(' ', '_').replace('/', '-')}_model_dashboard_part_1.png"
)

generate_model_driven_sponsor_dashboard(
    df,
    SPONSOR_NAME,
    save_path=f"{SPONSOR_NAME.replace(' ', '_').replace('/', '-')}_model_dashboard_part_2.png"
)
```

```python
# extract all unique sponsors, drop nulls, sort alphabetically
unique_sponsors = sorted(df["Sponsor"].dropna().unique().tolist())

print(f"Total unique sponsors in dataset: {len(unique_sponsors)}")
print()

# print them in a clean numbered format
for idx, sponsor in enumerate(unique_sponsors, start=1):
    print(f"{idx:>4}. {sponsor}")
```
